```python
from diffusers import StableDiffusionInpaintPipeline as StableDiffusionInpaintPipeline
```

```python
from io import BytesIO

from torch import autocast
import requests
import PIL
import torch
```

```python
from PIL import Image

def load_image(image_path):
    return Image.open(image_path).convert("RGB")
```

```python
img_path = "/content/drive/MyDrive/DIP_Project_Grp1/Images/person_dog.jpg"
mask_path = "/content/drive/MyDrive/DIP_Project_Grp1/Images/person_dog_mask.png"
```

```python
init_image = load_image(img_path).resize((512, 512))
mask_image = load_image(mask_path).resize((512, 512))
```

```python
device = "cuda"
pipe = StableDiffusionInpaintPipeline.from_pretrained(
    "CompVis/stable-diffusion-v1-4",
    revision="fp16",
    torch_dtype=torch.float16,
    use_auth_token=True
).to(device)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(init_image)
axes[0].set_title("Initial Image")
axes[0].axis('off')
axes[1].imshow(mask_image)
axes[1].set_title("Mask Image")
axes[1].axis('off')
plt.show()
```

```python
prompt = "a cat"
```

```python
with torch.autocast("cuda"):
    result = pipe(prompt=prompt, image=init_image, mask_image=mask_image, strength=0.7)
```

```python
images = result.images
```
